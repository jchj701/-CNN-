```python
def show_image_tensor(image_tensor):
    image = image_tensor.eval()
    print("图片的大小为：{}".format(image.shape))
    if len(image.shape) == 3 and image.shape[2] == 1:
        plt.imshow(image[:,:,0], cmap="Greys_r")
        plt.show()
    elif len(image.shape) == 3:
        plt.imshow(image)
        plt.show()
```

```python
def save_new_image(img_data, output_path, file, i):
    encoded_image = tf.image.encode_jpeg(img_data)
    file_full_write = os.path.join(output_path + '_new' + str(i) + '_' + file)
    with tf.gfile.GFile(file_full_write, "wb") as f:
        f.write(encoded_image.eval())
        print(file_full_write)
```

```python
def get_new_image(file, path_without_file):
    file_full_read = os.path.join(path_without_file + file)
    image_raw_data = tf.gfile.FastGFile(file_full_read,'rb').read()
    image_data = tf.image.decode_jpeg(image_raw_data, channels = 3)
    
    for i in range(5):
        procees_image(image_data, i,path_without_file, file)     
```

```python
def procees_image(image_raw_data, i, path_without_file, file):
    if i == 0:#上下翻转
        image_data = tf.image.flip_up_down(image_raw_data)
    elif i == 1:#左右翻转
        image_data = tf.image.flip_left_right(image_raw_data)
    elif i == 2:#旋转90°
        image_data = tf.image.transpose_image(image_raw_data)
    elif i == 3:#旋转-90°
        image_data = tf.image.rot90(image_raw_data, k=4)
    elif i == 4:#加入noise
        image_data = image_raw_data + tf.cast( tf.random_uniform(shape=[64, 64, 3],
                   minval=8,
                   maxval=12), tf.uint8)
    #show_image_tensor(image_data)#图片过多，不显示
    save_new_image(image_data, path_without_file, file, i)
```
